```python
df_2015['year']=2015
df_2016['year']=2016
df_2017['year']=2017
df_2018['year']=2018
df_2019['year']=2019

full_set = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019])
```

```python
from scikit_posthocs import posthoc_nemenyi_friedman

def nemenyi(col):
    cols = ['Unnamed: 0', 'year', col]
    subset = full_set[cols]
    subset.columns = ['config_name', 'year', col]
    return(pd.DataFrame(posthoc_nemenyi_friedman(subset, melted = True,  y_col = col, block_col = 'year', group_col = 'config_name')))
```

```python
resultsRMSE = nemenyi('RMSE')
resultsRMSE.to_excel('NemenyiRMSE.xlsx')

resultSPEARMAN = nemenyi('Spearman')
resultSPEARMAN.to_excel('NemenyiSpearman.xlsx')

resultACCWITHINONE = nemenyi('AccuracyWithinOne')
resultACCWITHINONE.to_excel('NemenyiAccuracyWithinOne.xlsx')

resultAVPREC = nemenyi('AUPRC')
resultAVPREC.to_excel('NemenyiAVPREC.xlsx')

resultRECALL = nemenyi('Recall')
resultRECALL.to_excel('NemenyiRecall.xlsx')
```
