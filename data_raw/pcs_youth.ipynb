```python
import requests
import pandas as pd

for race in races:
    for year in years:
        url = f'https://www.procyclingstats.com/race/{race}/{year}'
        response = requests.get(url)
        try:
            results = pd.read_html(response.text, match = 'Rider')[0]
        except ValueError: # Race results for that year not available
            continue
        results.assign(Race = race, Year = year).to_csv(f'data/{year}-{race}.csv', index = False)
```

```python
import os

files = os.listdir('data/')

all_results = pd.concat([pd.read_csv(f'data/{file}') for file in files])
all_results
```

```python
(all_results
 [['Rnk', 'Rider', 'Age', 'Team', 'UCI', 'Pnt', 'Time', 'Race', 'Year']]
 .assign(Rider = lambda x:
         x.assign(Team = all_results.Team.fillna('')).apply(lambda row: row.Rider.replace(row.Team, ''), axis = 1))
 .convert_dtypes()
 .fillna({'UCI': 0, 'Pnt': 0})
 .to_csv('pcs_youth_results.csv', index = False)
)
```
