```python
import pandas as pd

races_file_name = 'races.csv'
races = pd.read_csv(races_file_name)
races
```

```python
def load_race_results(race_id : int) -> pd.DataFrame:
    url = f'https://cqranking.com/men/asp/gen/race.asp?raceid={race_id}'
    results_df = pd.read_html(url, match = 'Time')[0].dropna(axis=1, how='all')
    results_df['race_id'] = race_id
    return results_df

race_cats_to_keep = ('1.WT1', '1.WT2', '1.WT3', '1.PS', '1.1', '1.2')
race_ids = races.loc[races.cat.isin(race_cats_to_keep)].race_id
result_dfs = (load_race_results(race_id) for race_id in race_ids)
results = pd.concat(result_dfs)
results
```

```python
results_file_name = 'results.csv'

(results
 .rename(columns = {'Rank': 'rank', 'Rider': 'rider', 'Team': 'team', 'Time': 'time', 'CQ': 'cq_points'})
 .to_csv(results_file_name, index = False)
)
pd.read_csv(results_file_name)
```
