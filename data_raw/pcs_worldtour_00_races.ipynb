```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def get_world_tour_races_by_year(year):
    print(year)
    url = f'https://www.procyclingstats.com/races.php?year={year}'
    response = requests.get(url)
    races = pd.read_html(response.text)[0]

    soup = BeautifulSoup(response.content, "html.parser")
    race_names = [a['href'].split('/')[1] for a in soup.table.find_all('a') if a['href'].startswith('race/')]
    return races.assign(
        race_name = race_names,
        year = year,
        stage_race = races.Date.str.contains(' - ')
    )[['race_name', 'year', 'stage_race']]

years = range(2013, 2023)
all_world_tour_races = [get_world_tour_races_by_year(year) for year in years]
pd.concat(all_world_tour_races).to_csv('world_tour_races.csv', index = False)
```
